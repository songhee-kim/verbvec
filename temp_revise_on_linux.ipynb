```python
def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result

#print(concatenate_list_data(['0', '1', 'M', 'M']))
```

```python
def syn_classify(arg_str, lemma):
    unerg_str = '^[^0-9]*0[^0-9]*'
    trans_str = '^[^0-9]*01[^0-9]*'
    unacc_str = '^[^0-9]*1[^0-9]*'
    if re.fullmatch(unacc_str, arg_str) != None:
        unacc_red.append(lemma)
    elif re.fullmatch(trans_str, arg_str) != None:
        trans_red.append(lemma)
    elif re.fullmatch(unerg_str, arg_str) != None:
        unerg_red.append(lemma)
```

```python
arg_config_dict = {} #dictionary containing argument configuration for all lemmas
unerg_red = []  #red means redundant
unacc_red = []
trans_red = []
error_lemma_red = []

attrib_name = [] 
attrib_name_passive = [] #list of attributes whose name starts with 'passiv'
lem_missing_example_name = []
inflection_voice_passive = [] #list of verbs whose inflection is marked as passive

#lemma_in_propbank2 = ['throw']
for lem in lemma_in_propbank:
    tree = ET.parse('/home/songhee/Projects/VerbVector/propbank/frames/%s.xml' % (lem))
    root = tree.getroot()
    predicate_set = root.findall('predicate')

    for pred in predicate_set:
        if pred.attrib['lemma'] == lem:
            example_set = pred.findall('roleset/example')
#             print (lem, len(example_set))
#             pass
            arg_config_per_lemma = []
            for es in example_set:
                infl_set = es.findall('inflection')
                if 'name' in es.keys():
                    #print (es.attrib['name'])
                    if ('passiv' in str(es.attrib['name'])) or ('-n' in str(es.attrib['name'])):
                        arg_config_per_lemma.append('NameFail')
                    else:
                        if len(infl_set)==0:
                            #attrib_name.append(str(es.attrib['name']))
                            arg_set = es.findall('arg')
#                             print (es.attrib['name'], arg_set)
#                             break
                            arg_per_example = []
                            for arg in arg_set:
                                #print (arg.get('f')=='', arg.attrib['n'])
                                if arg.get('f') == '':
                                    arg_per_example.append(arg.attrib['n'])
                            arg_str = concatenate_list_data(arg_per_example)
                            arg_config_per_lemma.append(arg_str)
                            syn_classify(arg_str, lem)
                        elif (len(infl_set)>0) and (infl_set[0].attrib['voice'] != 'passive'):
                            arg_set = es.findall('arg')
                            arg_per_example = []
                            for arg in arg_set:
                                if arg.get('f') == '':
                                    arg_per_example.append(arg.attrib['n'])
                            arg_str = concatenate_list_data(arg_per_example)
                            arg_config_per_lemma.append(arg_str)
                            syn_classify(arg_str, lem)
                        else:
                            arg_config_per_lemma.append('InflFail')
                            inflection_voice_passive.append(lem)
                     
                else: #'name' not in es.keys():
                    #print ('case 3', "name not found")
                    lem_missing_example_name.append(lem)
                    if len(infl_set)==0:
                        arg_set = es.findall('arg')
                        arg_per_example = []
                        for arg in arg_set:
                            if arg.get('f') == '':
                                arg_per_example.append(arg.attrib['n'])
                        arg_str = concatenate_list_data(arg_per_example)
                        arg_config_per_lemma.append(arg_str)                                
                        syn_classify(arg_str, lem)

                    elif (len(infl_set)>0) and (infl_set[0].attrib['voice'] != 'passive'):                        
                        arg_set = es.findall('arg')
                        arg_per_example = []
                        for arg in arg_set:
                            if arg.get('f') == '':
                                arg_per_example.append(arg.attrib['n'])
                        arg_str = concatenate_list_data(arg_per_example)
                        arg_config_per_lemma.append(arg_str)
                        #print (arg_per_example)
                        syn_classify(arg_str, lem)
                    else:
                        arg_config_per_lemma.append('NoExName-InflFail')

            arg_config_dict[lem] = arg_config_per_lemma
                            

    #arg_config_dict
#     except: 
#         print ("except", lem)
#         error_lemma_red.append(lem)
```
