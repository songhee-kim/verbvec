```python
def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result

#print(concatenate_list_data(['0', '1', 'M', 'M']))
```

```python
import re 

arg_config_dict = {} #dictionary containing argument configuration for all lemmas
unerg_red = []  #red means redundant
unacc_red = []
trans_red = []
error_lemma_red = []


for lem in lemma_in_propbank: 
    try:
        tree = ET.parse('/Users/songheekim/nltk_data/corpora/propbank/frames/%s.xml' % (lem))
        root = tree.getroot()
        predicate_set=root.findall('predicate')


        for i in range(len(predicate_set)):
            if predicate_set[i].attrib['lemma'] == lem:
                #print (predicate_set[i])
                example_set = predicate_set[i].findall('roleset/example')
                arg_config_list = []
                for j in range(len(example_set)):
                    arg_set = example_set[j].findall('arg')
                    #print (arg_set)
                    arg_per_example = []
                    for k in range(len(arg_set)):
                        arg_per_example.append(arg_set[k].attrib['n'])
                    arg_config_list.append(arg_per_example)
                    #print (arg_per_example)
                    
                    arg_str = concatenate_list_data(arg_per_example)
                    #print (arg_str)
                    
                    unerg_str = '^0[^0-9]*'
                    trans_str = '^[^0-9]*01[^0-9]*'
                    if arg_str.startswith('1'):
                        unacc_red.append(lem)
                        #print (lem, arg_str)
                    elif re.fullmatch(trans_str, arg_str) != None:
                        trans_red.append(lem)
                    elif re.fullmatch(unerg_str, arg_str) != None:
                        unerg_red.append(lem)


                arg_config_dict[lem] = arg_config_list


    #arg_config_dict
    except: 
        error_lemma_red.append(lem)
```
