```python
#classifiation based on 'example names'

corrupted_rolesets = []
unergative_list = []
unaccusative_list = [] 
roleset_example_dict = {}

roleset_name_redundant_list = []
for i in propbank.instances():
    roleset_name = i.roleset
    roleset_name_redundant_list.append(roleset_name)
roleset_name_list = list(set(roleset_name_redundant_list))


for roleset_name in roleset_name_list:
    verb = roleset_name.split('.')[0]
    if verb in lemma:         
        try: 
            examples = propbank.roleset(roleset_name).findall('example')
            example_n = len(examples)
            example_list = []
            for j in range(example_n):
                example_list.append(examples[j].attrib['name'])

            if roleset_name in roleset_example_dict.keys():
                pass
            else: 
                roleset_example_dict[roleset_name] = example_list

                
            if any('unaccusative' in example for example in example_list): 
                unaccusative_list.append(verb)
            elif any('unergative' in example for example in example_list):
                unergative_list.append(verb)
            elif any('intransitive' in example for example in example_list):
                #print (roleset_name, ' has intransitive in examples')
                index = example_list.index('intransitive')
                args = examples[index].findall('arg')
                args_n = len(args)
                arg_list = []
                for j in range(args_n):
                    arg_list.append(args[j].attrib['n'])
                print (verb, arg_list)

                if '0' in arg_list:
                    unergative_list.append(verb)
                elif '1' in arg_list:
                    unaccusative_list.append(verb)


        except: 
            corrupted_rolesets.append(roleset_name)
```

```python
unaccusative_verbs = sorted(list(set(unaccusative_list)))
```

```python
unergative_verbs = sorted(list(set(unergative_list)))
```
