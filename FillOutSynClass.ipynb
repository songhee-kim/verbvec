```python
#change SynClass (CELEX)

for k, v in rawdict.items():
    if v['SynClass(CELEX)'] == 'both':
        v['SynClass(CELEX)'] = 'ToBeDet'
    elif v['SynClass(CELEX)'] == 'intransitive':
        v['SynClass(CELEX)'] = 'Intransitive only'
    elif v['SynClass(CELEX)'] == 'transitive':
        v['SynClass(CELEX)'] = 'Transitive only'
```

```python
#update rawdict 
#if intrans only verb has 'Agent':1 & 'Theme':0 & 'Patient':0, 'SynClassVN'='Unergative'
#if intrans only verb has 'Agent':0 & ('Theme':1 or 'Patient':1), 'SynClassVN'='Unaccusative'
#if intrans only verb has 'Agent':1 & ('Theme':1 or 'Patient':1), print (intrans only, check)
#if trans-only verb has 'Agent':1, & ('Theme':1 or 'Patient:1), SynClass'='Transitive'

for k,v in rawdict.items():
    if (v['SynClass(CELEX)'] == 'Intransitive only') & (v['Agent'] ==1) & (v['Theme']==0) & (v['Patient']==0):
        v['SynClass(VN)'] = 'Unergative'
    elif (v['SynClass(CELEX)'] == 'Intransitive only') & (v['Agent'] ==0) & ((v['Theme']==1) or (v['Patient']==0)):
        v['SynClass(VN)'] = 'Unaccusative'
#     elif (v['SynClass(CELEX)'] == 'Intransitive only') & (v['Agent'] ==0) & (v['Theme']==0) & (v['Patient']==1):
#         v['SynClass(VN)'] = 'Unaccusative'
    elif (v['SynClass(CELEX)'] == 'Transitive only') & (v['Agent'] ==1) & (v['Theme']==1) & (v['Patient']==0):
        v['SynClass(VN)'] = 'Transitive'
    elif (v['SynClass(CELEX)'] == 'Transitive only') & (v['Agent'] ==1) & (v['Theme']==0) & (v['Patient']==1): 
        v['SynClass(VN)'] = 'Transitive'
```

```python
synadded = pd.DataFrame.from_dict(rawdict, orient='index')

#change the location of "SynClass (VN) column"
col_list = synadded.columns.tolist()
print ("the length of col_list is", len(col_list))
col_list.index('SynClassSource')
new_col_list = col_list[:8] + [col_list[-1]] + col_list[8:-1]
synadded2 = synadded[new_col_list]
```
